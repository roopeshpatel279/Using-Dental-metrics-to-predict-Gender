```python
# Drop unnecessary columns to create X (features)
X = df.drop(columns=["Gender"])

# Set Y (target variable) as "Gender"
Y = df["Gender"]

# Display the shapes of X and Y
X.shape, Y.shape
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data into training (70%) and testing (30%)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()
model_lr.fit(X_train,Y_train)
```

```python
model_lr.fit(X_train,Y_train)
print(model_lr.score(X_train,Y_train))
print(model_lr.score(X_test,Y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rfr=RandomForestClassifier(n_estimators=100,min_samples_split=25)
model_rfr.fit(X_train, Y_train)
print(model_rfr.score(X_train,Y_train))
print(model_rfr.score(X_test,Y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dtr=DecisionTreeClassifier(min_samples_split=23,min_samples_leaf=15)
model_dtr.fit(X_train, Y_train)
print(model_dtr.score(X_train,Y_train))
print(model_dtr.score(X_test,Y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_gbc=GradientBoostingClassifier()
model_gbc.fit(X_train, Y_train)
print(model_gbc.score(X_train,Y_train))
print(model_gbc.score(X_test,Y_test))
```

```python
# Reload necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df= pd.read_csv("Dentistry Dataset.csv")

# Set visualization style
sns.set_style("whitegrid")

# Figure setup
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# 1. Gender Distribution
sns.countplot(x="Gender", data=df, ax=axes[0, 0], palette="pastel")
axes[0, 0].set_title("Age")
axes[0, 0].set_xlabel("Gender")
axes[0, 0].set_ylabel("inter canine distance intraoral")

# 2. Age Distribution
sns.histplot(df["Age"], bins=20, kde=True, ax=axes[0, 1], color="lightblue")
axes[0, 1].set_title("Gender")
axes[0, 1].set_xlabel("Age")
axes[0, 1].set_ylabel("inter canine distance intraoral")

# 3. Intercanine Distance: Intraoral vs. Cast
sns.boxplot(data=df[["inter canine distance intraoral", "intercanine distance casts"]], ax=axes[1, 0], palette="muted")
axes[1, 0].set_title("Intercanine Distance: Intraoral vs. Cast")
axes[1, 0].set_xticklabels(["Intraoral", "Casts"])
axes[1, 0].set_ylabel("Measurement (mm)")

# 4. Canine Index Comparison by Gender
sns.boxplot(x="Gender", y="right canine index intra oral", data=df, ax=axes[1, 1], palette="pastel")
axes[1, 1].set_title("Right Canine Index by Gender")
axes[1, 1].set_xlabel("Gender")
axes[1, 1].set_ylabel("Right Canine Index")

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

features = ["inter canine distance intraoral", "left canine width intraoral", "right canine width intraoral"]
fig, axes = plt.subplots(len(features), 1, figsize=(10, 12), sharex=True)

for i, feature in enumerate(features):
    sns.kdeplot(df[df["Gender"] == "Male"][feature], shade=True, ax=axes[i], label="Male", color="blue")
    sns.kdeplot(df[df["Gender"] == "Female"][feature], shade=True, ax=axes[i], label="Female", color="red")
    axes[i].set_title(f"Distribution of {feature} by Gender")
    axes[i].legend()

plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt


# function for z axis
def f(x, y):
	return np.sin(np.sqrt(x ** 2 + y ** 2))

# x and y axis
x = np.linspace(-1, 5, 10)
y = np.linspace(-1, 5, 10)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot_wireframe(X, Y, Z, color ='green')
ax.set_title('variation in measurements');
```

```python
# Create a contour plot
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, cmap='viridis')
plt.colorbar(contour)
plt.title('Contour Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
